```python
import sqlite3
```

```python
connection = sqlite3.connect('example2.db')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
        create table employee(
        employee_id int primary key,
        name varchar(30),
        age int,
        salary float
        )
''')

connection.commit()
```

```python
cursor.execute('''
       select * from employee
''')
```

```python
cursor.execute('''
        insert into employee
        values
        (101,'sumit',20,20000),
        (102,'suryansh',23,40000)
''')

connection.commit()
```

```python
cursor.execute('Select * from employee')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
    update employee
    set age = 30
    where age < 25
''')

connection.commit()
```

```python
cursor.execute('Select * from employee')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
    delete from employee
    where
    name='suryansh'
''')

connection.commit()
```

```python
cursor.execute('Select * from employee')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## To insert multiple data

data = [
    (102,'shubham',20,40000),
    (103,'dhuriya',25,50000),
    (104,'amit',30,60000)
]

cursor.executemany('''
        insert into employee(employee_id,name,age,salary)
        values(?,?,?,?)
''',data)

connection.commit()
```

```python
cursor.execute('Select * from employee')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# To close the connection

connection.close()
```
